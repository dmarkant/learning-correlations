```python
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

## Generate datasets

```python
def generateDataset(rho,n=100):
    M = np.array([[1, rho],[rho, 1]])
    Y = np.random.multivariate_normal([0,0], M, size=n)
    Y[:,0] = Y[:,0]/np.std(Y[:,0] - Y[:,0].mean())
    Y[:,1] = Y[:,1]/np.std(Y[:,1] - Y[:,1].mean())
    return(Y)
```

```python
Y = generateDataset(.1, 100)
plt.plot(Y[:,0], Y[:,1], 'o')
```

```python
tol = .001 # sample correlations must be within .001 of target correlation
d_rho = .1 # step size of correlations
iterations = 100      # number of datasets per combination

for ind in range(-10, 11, 1):
    
    rho = ind/10.
    
    for n in [10, 50, 100]:
        for ind in range(iterations):

            err = 1
            while err > tol:

                Y = generateDataset(rho, n)
                r = pearsonr(Y[:,0], Y[:,1])[0]
                err = np.abs(rho - r)

            print(rho, n, ind)
            np.savetxt(f'datasets/dataset_r={np.round(rho,2)}_n={n}_ind={ind}.txt', np.round(Y, 2), '%1.2f')
    
```
